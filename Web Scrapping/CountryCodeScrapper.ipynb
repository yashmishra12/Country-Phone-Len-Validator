```python
html_tables = pd.read_html(URL)
```

```python
df = html_tables[0]
```

```python
df.iloc[0]
```

```python
returner = {'Country': [], 'Code': [], 'ISO':[]}
for i in range (0, len(df)):
    returner['Country'].append(df.iloc[i]["COUNTRY"])
    returner['Code'].append(df.iloc[i]["COUNTRY CODE"])
    returner['ISO'].append(df.iloc[i]["ISO CODES"])

returner_df = pd.DataFrame(returner, columns=['Country','Code','ISO']) 
```
